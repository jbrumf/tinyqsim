## Prototype of Bloch Sphere

This is still under development, so is likely to change.

````python
PI = '\u03C0'  # PI unicode
from math import pi
from tinyqsim.qcircuit import QCircuit
from tinyqsim.bloch import plot_bloch
````

````python
qc = QCircuit(1)  # This must be 1 qubit at present
qc.h(0)
qc.t(0)
qc.draw()
print(qc.state_vector)
plot_bloch(qc.state_vector, scale=1.0)
````

````python
qc = QCircuit(1)  # This must be 1 qubit at present
qc.h(0)
qc.p(pi/3, f'{PI}/3', 0)
qc.draw()
print(qc.state_vector)
plot_bloch(qc.state_vector)
````

### Notes

The sphere has a fixed orientation when displayed within a Jupyter Notebook. The view point can be set using the optional parameters azimuth and elevation in degrees:
```
  plot_bloch(psi, azimuth=35, elevation=10)
```
